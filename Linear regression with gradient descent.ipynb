```python
#making the inputs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12.0,9.0)
```

```python
# Preprosessing the input data
data=pd.read_csv('data2.txt')
X=data.iloc[:,0]
Y=data.iloc[:,1]
plt.scatter(X,Y)
plt.show()
```

```python
#building the model
w0 = 0
w1 = 0
alpha = 0.01
iteration = 1500
m = float(len(X))

# gradient descent
for i in range(iteration):
    predicted_value= w0 + w1*X #the current predicted value
    d_w0 = (1/m) * sum(predicted_value-Y) 
    d_w1 = (1/m) * sum((predicted_value-Y)*X)
    w0 = w0 - ((alpha/m)*d_w0)
    w1 = w1 - ((alpha/m)*d_w1)
    
print(w0,w1)
```

```python
#making predictions
predicted_value= w0 + w1*X
plt.scatter(X,Y)
plt.plot([min(X),max(X)],[min(predicted_value),max(predicted_value)],color='red')
plt.show()
```
